# Impoting Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading Tables

```python

games = pd.read_csv("games.csv")   
recommendations = pd.read_csv("recommendations.csv")
games_metadata = pd.read_json("games_metadata.json",lines=True) 
```

# Understanding Data

### < Checking Columns >

```python
print("games Table Columns Name",list(games.columns))
print("**"*50)
print()
print("recommendations Table Columns Name",list(recommendations.columns))
print("**"*50)
print()
print("games_metadata Table Columns Name",list(games_metadata.columns))
```

```python
recommendations.rename(columns = {"date":"Date_of_publishing"},inplace=True)
```

### < Merging Tables Together >

```python
recommendations_Games_df = pd.merge(recommendations,games,on="app_id")
recommendations_Games_df.drop_duplicates(inplace=True)
recommendations_Games_df
```

#  Checking Columns Data Types

```python
recommendations_Games_df.info()
```

```python
recommendations_Games_df["Date_of_publishing"] = pd.to_datetime(recommendations_Games_df["Date_of_publishing"],format="%Y-%m-%d")
```

```python
recommendations_Games_df["date_release"] = pd.to_datetime(recommendations_Games_df["date_release"],format="%Y-%m-%d")
```

```python
recommendations_Games_df.info()
```

# EDA Questions


### 1)Top 10 Games With Rating 'Very Positive'

```python
recommendations_Games_df[recommendations_Games_df['rating']=='Very Positive' ].groupby('title')[['rating']].value_counts()
                                                         .sort_values(ascending=False).head(10).reset_index()
```

### 2)Top 10 Games With Rating 'Very Negative'

```python
recommendations_Games_df[recommendations_Games_df['rating']=='Very Negative' ].groupby('title')[['rating']].value_counts()
                                                                   .sort_values(ascending=False).head(10).reset_index()
```

### 3) Max Hours may played by user

```python
recommendations_Games_df['hours'].max()
```

### 4) Top 10 games with much Hours played by users

```python
recommendations_Games_df[recommendations_Games_df['hours'] == recommendations_Games_df['hours'].max()]
                                                                            [['title','hours','user_id']].head(10)
```

### 5) Min Hours may played by user

```python
recommendations_Games_df['hours'].min()
```

### 6) 10 games with least Hours played by users

```python
recommendations_Games_df[recommendations_Games_df['hours'] == recommendations_Games_df['hours'].min()][['title','hours','user_id']].head(10)
```

# 7) Most Expansive 10 Games

```python
price_game = games[['title','price_final']]
price_game.drop_duplicates().sort_values(by='price_final',ascending=False).head(10).reset_index()
```

### 8) Cheapest 10 Games

```python
price_game.drop_duplicates().sort_values(by='price_final',ascending=True).head(10).reset_index()
```

### 9) top ten most Recomended games by users

```python
rec_games = recommendations_Games_df[['helpful','title']] 
rec_games.drop_duplicates().sort_values(by='helpful',ascending=False).head(10).reset_index()
```
